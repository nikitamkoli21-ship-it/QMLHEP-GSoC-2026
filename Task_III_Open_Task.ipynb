## Understanding Quantum Machine Learning (QML)
During Task I, I understood that QML is not a drastic change from classical Machine Learning, rather it is one step further in its development. For example, instead of neural network layers with weights and biases, we create Quantum Circuits. The "trainable parameters" in this setting are the rotation angles of quantum gates (like $RX, RY, and RZ$). While using PennyLane, I discovered that we can optimize these angles via gradient descent in the same way we do it in PyTorch. This hybrid method employing classical optimizers in the tuning of quantum circuits can serve as an effective introduction to the NISQ (Noisy Intermediate-Scale Quantum) era.
