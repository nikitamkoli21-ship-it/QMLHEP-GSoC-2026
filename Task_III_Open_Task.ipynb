# Task III — My Thoughts on Quantum Machine Learning

## Quantum Computing — What I Understood
Normal computers that we use every day work with bits.
A bit is either 0 or 1, nothing else. But quantum computers
are different — they use something called qubits.

What makes qubits special is that they don't have to be
just 0 or 1. They can be both at the same time! This is
called superposition. I think of it like a coin that is
spinning — while it spins, it's both heads and tails.
Only when it lands (when we measure it) do we get a
definite answer.

Another thing I found interesting is entanglement.
If two qubits are entangled, whatever happens to one
immediately affects the other — even if they are far apart.
This is what Einstein called "spooky action at a distance."

## Quantum Machine Learning — My Understanding
After doing Task I with PennyLane, I started to understand
what Quantum ML actually means. Instead of building a
neural network with layers and weights like in PyTorch,
in QML we build a quantum circuit with rotation angles
as the trainable parameters.

These rotation angles (like RX, RY, RZ that I used in Task I)
get updated during training — exactly like weights in a
normal neural network. So the idea is not completely
different from classical ML, just the hardware is different.

## Algorithm I Learned — Variational Quantum Circuit (VQC)
The most interesting algorithm I came across while
preparing for this project is the Variational Quantum
Circuit or VQC.

Here is how I understand it:
- First, we take classical data (like a number or image)
  and encode it into a quantum state
- Then we apply gates with trainable rotation angles
- Then we measure the output of the circuit
- We calculate the loss and update the angles using
  gradient descent

This is basically a quantum version of a neural network.
What excites me about VQCs is that they might be able to
find patterns in data that classical computers struggle with.
